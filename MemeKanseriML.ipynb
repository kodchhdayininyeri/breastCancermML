```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataFrame = pd.read_csv("breast_cancer.csv")
```

```python
dataFrame
```

```python
x = dataFrame.iloc[:,1:-1].values

y = dataFrame.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
#train and test set sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# set logistic regression model
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
```

```python
#fit
classifier.fit(x_train,y_train)
```

```python
#test model
y_pred = classifier.predict(x_test)
```

```python
y_pred_prob=classifier.predict_proba(x_test)
```

```python
import seaborn as sns

df = pd.DataFrame(y_pred_prob, columns=["Class 0", "Class 1"])

plt.figure(figsize=(10, 50))
sns.heatmap(df, annot=True, cmap="RdBu")
plt.title("Sample Data Heatmap")
plt.show()
```

```python
y_pred
```

```python
#Confusion Matrix and Accuracy Score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
accuracyscore=accuracy_score(y_test,y_pred)
```

```python
accuracyscore
```

```python
#we measure model performance with k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuricies = cross_val_score(estimator=classifier,X=x,y=y,cv=10)
```
